```python
import pandas as pd
import numpy as np
df = pd.read_csv("housing.csv")

df["income_cat"] = pd.cut(df["median_income"],
                             bins=[0, 1.5, 3.0, 4.5, 6.0, np.inf],    # creates income_cat and divides it into catoagroies so it can be easily split for test and train data
                             labels=[1, 2, 3, 4, 5])
```

```python
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
                                                                                #stratified shuffle using income_cat , so that both  test and train data have all income cat 
for train_index, test_index in split.split(df, df["income_cat"]):
    strat_train_set = df.loc[train_index]
    strat_test_set = df.loc[test_index]
```

```python
#lets remove income_cat coloumn
for sett in(strat_train_set, strat_test_set):
    sett.drop("income_cat",axis =1,inplace = True)
```

```python
df = strat_train_set.copy()
```
